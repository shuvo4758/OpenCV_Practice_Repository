# Drawing Line

```python
# cv2.line(image, point1, point2, color, thickness)
# point1 = x1, y1
# point2 = x2, y2
# color = (B, G, R)

import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread('Goku.jpg')

if image is not None:
  print('Image loaded')

  point1 = (1000, 200)
  point2 = (2000, 2500)
  color = (0, 0, 255)
  thickness = 10
  cv2.line(image, point1, point2, color, thickness)

  cv2_imshow(image)
  cv2.waitKey(0)
  cv2.destroyAllWindows()
else:
  print('Image not loaded')
```

# Drawing Rectangle

```python
# cv2.rectangle(image, point1, point2, color, thickness)
# point1 = x1, y1 --> top left corner
# point2 = x2, y2 --> bottom right corner

import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread('Goku.jpg')

if image is not None:
  print('Image loaded')

  point1 = (1000, 200)
  point2 = (3000, 1500)
  color = (0, 0, 255)
  thickness = 20    # if -1 --> solid rectangle filled with that color
  cv2.rectangle(image, point1, point2, color, thickness)

  cv2_imshow(image)
  cv2.waitKey(0)
  cv2.destroyAllWindows()
else:
  print('Image not loaded')
```

# Drawing Circle

```python
# cv2.circle(image, center, radius, color, thickness)

import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread('Goku.jpg')

if image is not None:
  print('Image loaded')

  center = (1900, 1000)
  radius = 700
  color = (0, 0, 255)
  thickness = 20
  cv2.circle(image, center, radius, color, thickness)

  cv2_imshow(image)
  cv2.waitKey(0)
  cv2.destroyAllWindows()
else:
  print('Image not loaded')
```

# Adding Text

```python
# cv2.putText(image, text, org, font, fontScale, color, thickness)
# org = bottom left corner of the text

import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread('Goku.jpg')

if image is not None:
  print('Image loaded')

  text = 'SuperSaiyan Goku'
  org = (1000, 400)
  font = cv2.FONT_HERSHEY_SIMPLEX
  fontScale = 5
  color = (0, 255, 255)
  thickness = 10
  cv2.putText(image, text, org, font, fontScale, color, thickness)

  cv2_imshow(image)
  cv2.waitKey(0)
  cv2.destroyAllWindows()
else:
  print('Image not loaded')
```

# Mini Project using all Drawings

```python
import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread(input('Enter your Image location:'))

print('What do you want to Draw?')
print('1. Line \n2. Rectangle \n3. Circle \n4. Text')
choice = input('Enter your choice:')

def points(choice):
    point1 = (int(input('Enter x1:')), int(input('Enter y1:')))
    point2 = (int(input('Enter x2:')), int(input('Enter y2:')))
    colorThickness(point1, point2, choice)

def colorThickness(point1, point2, choice):
      color = (int(input('Enter B:')), int(input('Enter G:')), int(input('Enter R:')))
      thickness = int(input('Enter thickness:'))

      if choice == '1' or choice.lower() == 'line':
        cv2.line(image, point1, point2, color, thickness)
        display()

      elif choice == '2' or choice.lower() == 'rectangle': # Fixed: choice.lower(choice) to choice.lower()
        cv2.rectangle(image, point1, point2, color, thickness)
        display()

      elif choice == '3' or choice.lower() == 'circle':
        cv2.circle(image, point1, point2, color, thickness)
        display()

      elif choice == '4' or choice.lower() == 'text':
        # Fixed: Unpack text, org, font, and fontScale from point1 and point2
        text, org = point1
        font, fontScale = point2
        cv2.putText(image, text, org, font, fontScale, color, thickness)
        display()

      else:
        print('Invalid Choice')


def display():
  cv2_imshow(image)
  cv2.waitKey(0)
  cv2.destroyAllWindows()

if image is not None and (choice == '1' or choice.lower() == 'line'):
  points(choice)

elif image is not None and (choice == '2' or choice.lower() == 'rectangle'):
  points(choice)

elif image is not None and (choice == '3' or choice.lower() == 'circle'):
  center = (int(input('Enter center of x:')), int(input('Enter center of y:')))
  radius = int(input('Enter radius:'))
  colorThickness(center, radius, choice)

elif image is not None and (choice == '4' or choice.lower() == 'text'):
  text = input('Enter your text:')
  org = (int(input('Enter x:')), int(input('Enter y:')))
  font = cv2.FONT_HERSHEY_SIMPLEX
  fontScale = int(input('Enter fontScale:'))
  text_Org = (text, org)
  font_Fontscale = (font, fontScale)
  colorThickness(text_Org, font_Fontscale, choice)

else:
  print('Invalid Choice')
```
